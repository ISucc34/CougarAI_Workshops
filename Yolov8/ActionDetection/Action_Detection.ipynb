## Training a Model

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")

model.train(data="Punch_Data/data.yaml", epochs = 50)
```

```python
#Training through the command line

#Terminal Training
#!yolo task=detect mode=train data=data.yaml model=yolov8n.pt epochs=50
```

##Validating the trained model


```python
#Through Code

model = YOLO("/content/runs/detect/train/weights/best.pt")

results = model.val()
```
