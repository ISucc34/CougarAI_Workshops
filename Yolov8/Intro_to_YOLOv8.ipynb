#Base models using code

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt") #take a pretrained model
model.predict(source = "dog.jpeg", save=True) #predict using a source
```

```python
from numpy import true_divide
model = YOLO("yolov8n-seg.pt")
model.predict(source = "dog.jpeg", save=True)
```
