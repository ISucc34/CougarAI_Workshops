```python
from ultralytics import YOLO
```

## Training the Model

```python
model = YOLO('yolov8n.pt')
path = "/content/MMA Striking/data.yaml"

model.train(data=path, epochs = 50, batch = 10)
```

## Validation

```python
model = YOLO('/content/runs/detect/train5/weights/best.pt')
model.val()
```
